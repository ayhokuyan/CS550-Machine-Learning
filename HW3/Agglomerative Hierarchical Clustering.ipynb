```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg 
import time
from scipy.spatial import distance
```

```python
class KMeans():
    def __init__(self, data, K=2, normalize=True):
        #preserve dimension information 
        self.shape = data.shape
        self.features = data.shape[-1]
        
        #save the flattened data as data (K,M,3) -> (KxM,3)
        self.data = np.reshape(data, (-1,self.features))

        
        #save number of clusters
        self.K = K
        
        # Used Forgy's Method to initialize, meaning we selected random number of data points 
        #from the data as cluster centroids.
        ind = np.random.permutation(self.data.shape[0])
        self.centroids = np.array(self.data[ind][:K,:], copy=True)
        
        # initialize membership information for each data point
        self.membership = 0
        
    def expectation(self):
        distances = np.sqrt(np.sum(((self.data - self.centroids[:,np.newaxis])**2), axis=-1))
        self.membership = np.argmin(distances, axis=0)

    def maximization(self):
        error = 0
        for i in range(self.K):
            self.centroids[i,:] = np.mean(self.data[self.membership == i], axis=0)
            #mean squared error is averaged through all pixels
            error += np.sum(np.sqrt(np.sum((self.data[self.membership == i] - self.centroids[i,:])**2, axis=-1)))
        error /= self.data.shape[0]
        #print(error)
        #print(self.centroids)
        return error
        
    def train(self, error_margin=0.001, verbose=False):
        history = dict()
        errors = [np.inf]
        start_time = time.time()
        while(True):
            self.expectation()
            error = self.maximization()
            if verbose:
                print(error)
            if(np.abs(error - errors[-1]) <= error_margin):
                break;
            errors.append(error)
        stop_time = time.time()
        if verbose:
            print('Elapsed time: %.5fs' % (stop_time - start_time))
            
        return self.centroids, self.membership
    
    def visualize(self):
        img = np.ones(self.data.shape)
        for i in range(self.K):
            img[self.membership == i] = self.centroids[i]
        # if data normalized, return it to its original mean and standard deviation
        if(self.normalize):
            img *= self.std
            img += self.mean
        img = np.round(img).astype(int)
        img = img.reshape(self.shape)
        return img
    
    def getCentroids(self):
        return self.centroids*self.std +self.mean
```

As we implemented k-means already, we will first use it to segment the image to small clusters, such as K=100, then use Agglomerative Clustering to further segmentation. 

```python
class HACwKMeans():
    def __init__(self, data, K_final=2, K_init=100, normalize=True):
        #preserve dimension information 
        self.shape = data.shape
        self.features = data.shape[-1]
        
        #save the flattened data as data (K,M,3) -> (KxM,3)
        self.data = np.reshape(data, (-1,self.features))
        
        #normalize data
        self.mean = np.mean(self.data)
        self.std = np.std(self.data)
        self.normalize = normalize
        if(normalize):
            self.data = (self.data - self.mean)/self.std
        
        #save number of clusters
        self.K_final = K_final
        self.K_init = K_init
        
        #initialize centroids and memberships
        self.centroids = 0
        self.membership = 0
        
        #init klist and the distance matrix
        self.k = 0
        self.klist = 0
        self.distanceMatrix = 0
        
    def calcDistance(self, c1, c2, linkage):
        #print(c1.shape, c2.shape)
        distances = distance.cdist(c1, c2, 'euclidean')
        #print(distances.shape)
        if(linkage == 'single'):
            return np.min(distances)
        elif(linkage == 'complete'):
            return np.max(distances)
        else:
            raise ValueError('linkage can only be single or complete')
        
    
    def calcDistanceMatrix(self,linkage, verbose):
        distances = np.zeros((self.k,self.k))
        for i in range(self.k):
            for j in range(self.k):
                if verbose:
                    print('Distance (%d,%d)' % (i,j))
                if(i != j and distances[i,j] == 0 and distances[j,i] == 0):
                    dist = self.calcDistance(self.data[self.membership == j],
                                             self.data[self.membership == i],
                                             linkage)
                    distances[i,j] = dist
                    distances[j,i] = dist
                elif(distances[i,j] != 0):
                    distances[j,i] = distances[i,j]
                elif(distances[j,i] != 0):
                    distances[i,j] = distances[j,i]
        self.distanceMatrix = distances
        
    def iterate(self, linkage):
        indices = np.argwhere(self.distanceMatrix == np.amin(self.distanceMatrix[self.distanceMatrix!=0]))[0]
        ind1 = indices[0]
        ind2 = indices[1]
        #print('ind',ind1,ind2)
        #print('k_ind', self.klist[ind1],self.klist[ind2])    
        
        #reassign memberships of 2nd cluster
        self.membership[self.membership == self.klist[ind2]] = self.klist[ind1]
        
        #set the new distances depending on the linkage
        if linkage == 'single':
            self.distanceMatrix[ind1,:] = np.minimum(self.distanceMatrix[ind2,:],self.distanceMatrix[ind1,:])
            self.distanceMatrix[:,ind1] = np.minimum(self.distanceMatrix[:,ind2],self.distanceMatrix[:,ind1])
            self.distanceMatrix[ind1,ind1] = 0 
        else:
            self.distanceMatrix[ind1,:] = np.maximum(self.distanceMatrix[ind2,:],self.distanceMatrix[ind1,:])
            self.distanceMatrix[:,ind1] = np.maximum(self.distanceMatrix[:,ind2],self.distanceMatrix[:,ind1])
            self.distanceMatrix[ind1,ind1] = 0 
        #delete the row and column of 2nd cluster
        self.distanceMatrix = np.delete(self.distanceMatrix, ind2, axis=0)
        self.distanceMatrix = np.delete(self.distanceMatrix, ind2, axis=1)
        
        #delete ind2 from cluster list
        self.k -= 1
        removed = self.klist[ind2]
        self.klist = self.klist[self.klist != removed]
        
        #print('dm', self.distanceMatrix.shape)
        #print(np.unique(self.membership, return_counts=True))
        #print('klist', self.klist)
        return removed
    
    def setCentroids(self):
        self.centroids = np.ones((self.K_final, self.features))
        for x,i in enumerate(self.klist):
            self.centroids[x,:] = np.mean(self.data[self.membership == i], axis=0)
            
    def calcError(self):
        error = 0
        for i,cl in enumerate(self.klist):
            #mean squared error is averaged through all pixels
            error += np.sum(np.sqrt(np.sum((self.data[self.membership == cl] - self.centroids[i,:])**2, axis=-1)))
        error /= self.data.shape[0]
        return error
            
        
    def train(self, error_margin=0.001, verbose=False, linkage='single'):
        history = dict()
        
        #train a k-means algorithm to find initial clusters and set them
        start_time = time.time()
        kmc = KMeans(self.data, K=self.K_init, normalize=False)
        self.centroids, self.membership = kmc.train(error_margin, verbose)
        print('Trained k-Means on data with K=%d for %.5fs' % (self.K_init,time.time()-start_time))
        self.k = self.K_init
        self.klist = np.unique(self.membership)
        
        ##TRAIN CODE##
        print('Setting up the initial distances...')
        self.calcDistanceMatrix(linkage, verbose)
        print('Done, time passed= %.5fs' % (time.time()-start_time) )
        
        
        #print('dm', self.distanceMatrix.shape)
        #print(np.unique(self.membership, return_counts=True))
        #print('klist', self.klist)
        
        while(self.k > self.K_final):
            rem = self.iterate(linkage)
            if verbose:
                print('%d merged, %d clusters remain' % (rem, self.k))
        
        stop_time = time.time()
        print('Elapsed time: %.5fs' % (stop_time - start_time))
        
        #set new centroids
        self.setCentroids()
        
        history['time'] = (stop_time - start_time)
        history['clustering_error'] = self.calcError()
        history['centroids'] = self.getCentroids()
            
        return history
    
    def visualize(self):
        img = np.ones(self.data.shape)
        
        for x,i in enumerate(self.klist):
            img[self.membership == i] = self.centroids[x]
        #print(np.unique(img))
        # if data normalized, return it to its original mean and standard deviation
        if(self.normalize):
            img *= self.std
            img += self.mean
        img = np.round(img).astype(int)
        img = img.reshape(self.shape)
        return img
    
    def getCentroids(self):
        return self.centroids*self.std +self.mean
```

```python
#Choosing between single and complete linkage
leak = ['single', 'complete']
images = list()
for lk in leak:
    cl = HACwKMeans(data, K_init=100, K_final=10)
    cl.train(error_margin=0.0000, linkage=lk, verbose=False)
    images.append(cl.visualize())
plt.figure(figsize=(20,20))
for i,img in enumerate(images):
    plt.subplot(1,2,i+1)
    plt.imshow(img)
    plt.axis('off')
    plt.title('%s linkage' % leak[i])
plt.show()
```

```python
def formatResults(data, KList, n, m):
    times = []
    errors = []
    cluster_vectors =list()
    images = list()
    
    for k in KList:
        cl = HACwKMeans(data, K_init=100, K_final=k)
        history = cl.train(error_margin=0.00, linkage='complete', verbose=False)
        
        
        times.append(history['time'])
        errors.append(history['clustering_error'])
        cluster_vectors.append('centroids')
        
        images.append(cl.visualize())
        
    
    plt.figure(figsize=(m*4,n*4))
    for i in range(len(KList)):
        plt.subplot(n,m,i+1)
        plt.imshow(images[i])
        plt.title('K={}'.format((KList[i])))
        plt.axis('off')
    plt.show()
    
    print('Execuction Times:\n', times)
    print('Errors:\n', errors)

Ks = [2,3,4,5,6,10]
formatResults(data, Ks, 3, 2)
```
