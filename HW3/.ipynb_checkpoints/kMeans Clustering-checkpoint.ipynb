```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg 
import time
```

```python
data = mpimg.imread('sample.jpg')
plt.imshow(data)
plt.title('Sample.jpg')
plt.axis('off')
plt.show()
print(data.shape)
vls,cts = np.unique(data.reshape(-1,3), return_counts=True, axis=0)
print(len(cts))
```

```python
class KMeans():
    def __init__(self, data, K=2, normalize=True):
        #preserve dimension information 
        self.shape = data.shape
        self.features = data.shape[-1]
        
        #save the flattened data as data (K,M,3) -> (KxM,3)
        self.data = np.reshape(data, (-1,self.features))
        
        #normalize data
        self.mean = np.mean(self.data)
        self.std = np.std(self.data)
        self.normalize = normalize
        if(normalize):
            self.data = (self.data - self.mean)/self.std
        
        #save number of clusters
        self.K = K
        
        # Used Forgy's Method to initialize, meaning we selected random number of data points 
        #from the data as cluster centroids.
        ind = np.random.permutation(self.data.shape[0])
        self.centroids = np.array(self.data[ind][:K,:], copy=True)
        
        # initialize membership information for each data point
        self.membership = 0
        
    def expectation(self):
        distances = np.sqrt(np.sum(((self.data - self.centroids[:,np.newaxis])**2), axis=-1))
        self.membership = np.argmin(distances, axis=0)

    def maximization(self):
        error = 0
        for i in range(self.K):
            self.centroids[i,:] = np.mean(self.data[self.membership == i], axis=0)
            #mean squared error is averaged through all pixels
            error += np.sum(np.sqrt(np.sum((self.data[self.membership == i] - self.centroids[i,:])**2, axis=-1)))
        error /= self.data.shape[0]
        #print(error)
        #print(self.centroids)
        return error
        
    def train(self, error_margin=0.001, verbose=False):
        history = dict()
        errors = [np.inf]
        start_time = time.time()
        while(True):
            self.expectation()
            error = self.maximization()
            if verbose:
                print(error)
            if(np.abs(error - errors[-1]) <= error_margin):
                break;
            errors.append(error)
        stop_time = time.time()
        if verbose:
            print('Elapsed time: %.5fs' % (stop_time - start_time))
            
        history['error'] = errors[1:]
        history['time'] = stop_time - start_time
        history['clustering_error'] = (errors[-1])
        return history
    
    def visualize(self):
        img = np.ones(self.data.shape)
        for i in range(self.K):
            img[self.membership == i] = self.centroids[i]
        # if data normalized, return it to its original mean and standard deviation
        if(self.normalize):
            img *= self.std
            img += self.mean
        img = np.round(img).astype(int)
        img = img.reshape(self.shape)
        return img
    
    def getCentroids(self):
        return self.centroids*self.std +self.mean
```

```python
def formatResults(data, KList, n, m):
    times = []
    errors = []
    cluster_vectors =list()
    images = list()
    
    for k in KList:
        cl = KMeans(data, K=k)
        history = cl.train(error_margin=0.00001)
        times.append(history['time'])
        errors.append(history['clustering_error'])
        
        cluster_vectors.append(cl.getCentroids())
        images.append(cl.visualize())
        
    
    plt.figure(figsize=(m*4,n*4))
    for i in range(len(KList)):
        plt.subplot(n,m,i+1)
        plt.imshow(images[i])
        plt.title('K={}'.format((KList[i])))
        plt.axis('off')
    plt.show()
    
    print('Execuction Times:\n', times)
    print('Errors:\n', errors)
    
    for i in range(len(KList)):
        print('Vectors for K=%d' % KList[i])
        print(np.around(cluster_vectors[i], decimals=2))

Ks = [2,3,4,5,6,10]
formatResults(data, Ks, 3, 2)
```

```python
def findBestK(data, KList):
    errors = list()
    for k in KList:
        cl = KMeans(data, K=k)
        history = cl.train(error_margin=0.00001)
        errors.append(history['clustering_error'])
    plt.plot(KList, errors)
    plt.title('Clustering Error vs K')
    plt.xlabel('K')
    plt.ylabel('Normalized Clustering Error')
    plt.show()

Ks = [2,3,4,5,6,8,10,16,32,64,100,120,200]
findBestK(data, Ks)
```

We decided that K=30 is a good threshold, after that as the error gets lower more slowly, the execution time increases. 

```python
BEST_K =30
cl = KMeans(data, K=BEST_K)
history = cl.train(error_margin=1e-6, verbose=True)
```

```python
print('Clustering Error:', history['clustering_error'])

image = cl.visualize()
plt.imshow(image)
plt.axis('off')
plt.title('Clustered Image with K=30')
plt.show()
```
