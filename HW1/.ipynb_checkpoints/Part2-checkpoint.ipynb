```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
```

```python
#load train_data
#set current working directory
cwd = os.getcwd()
#load train data as csv
train_data = pd.read_csv(os.path.join(cwd, 'train_data.txt'), sep=' ', header=None)
train_data.reset_index(drop=True, inplace=True)
train_data = train_data.to_numpy()[:,:-2]

#load test data as csv
test_data = pd.read_csv(os.path.join(cwd, 'test_data.txt'), sep=' ', header=None)
test_data.reset_index(drop=True, inplace=True)
test_data = test_data.to_numpy()[:,:-2]

#convert to features and labels
print(train_data.shape)
print(test_data.shape)

numFeatures = train_data.shape[1]
train_features = train_data[:,:numFeatures-2]
train_labels = train_data[:,numFeatures-1]
test_features = test_data[:,:numFeatures-2]
test_labels = test_data[:,numFeatures-1]

print(train_features.shape)
print(train_labels.shape)

#load feature names 
featNames = pd.read_csv(os.path.join(cwd, 'thy_cost.txt'), sep=':', header=None)
#featNames.reset_index(drop=True, inplace=True)
featNames = featNames.to_numpy()[:,0]
print(featNames, featNames.shape)
featNames = np.append(featNames, '-')
```

```python
def gini_calc(node_labels):
    labels, counts = np.unique(node_labels, return_counts=True)
    gini = 0
    all_counts = np.sum(counts)
    for i in range(len(labels)):
        gini += (counts[i]/all_counts)**2 
    gini_imp = (1 - gini)
    return gini_imp   

def bestSplitofFeature(feature, labels):
    best_wimp = 2
    threshold = np.inf
    for val in feature:
        left_split = label[labels <= val]
        p_left = len(left_split)/len(labels)
        right_split = label[labels > val]
        p_right = len(right_split)/len(labels)
        weighted_impurity = p_left * gini_calc(left_split) + (1-p_left) * gini_calc(right_split)
        if(weighted_impurity <= best_wimp):
            best_wimp = weighted_impurity;

def best_split(node_features, node_labels):
    numFeatures = node_features.shape[1] 
    #print(numFeatures)
    wimps = []
    split_vals = []
    split_indices = []
    sort_indices = []
    
    for i in range(numFeatures):
        split_feature = node_features[:,i]
        #print('feature:', split_feature)
        split_val = 1
        split_index = 20
        weighted_impurity = np.inf
        
        
        all_counts = len(node_labels)
        
        #sort sample for the speicifc feature
        sort_index = np.lexsort((node_labels,split_feature))
        split_feature = split_feature[sort_index]
        node_labels_feat = node_labels[sort_index]
        
        #check if binary. If it is, there's only one split. 
        if(np.array_equal(split_feature, split_feature.astype(bool))):
            #binary
            #find first occurence of one, (there is a high probabiility of zeros), then i-1 is the index of last 0 elemet. 
            #print(np.where(split_feature == 1)[0])

        
            locZeros = np.where(split_feature == 0)[0]
            if(len(locZeros) == 0 or len(locZeros) == len(split_feature)):
                #ineligible indexes and values, doesn't matter since weighted impurity will be set to 1
                split_index = -1
                weighted_impurity = np.inf
                split_val = -1
            else:
                split_index =  locZeros[-1] 
                left_labels = node_labels_feat[:split_index+1]
                right_labels = node_labels_feat[split_index+1:]
            
                p_left = (split_index+1)/all_counts
                weighted_impurity = p_left * gini_calc(left_labels) + (1-p_left) * gini_calc(right_labels)
                split_val = split_feature[split_index]            
            
        else:
            #continuous
            for split in range(len(split_feature)-1):
                left_labels = node_labels_feat[:split+1]
                right_labels = node_labels_feat[split+1:]
                #print(left_labels, right_labels)
                p_left = (split+1)/all_counts
                weighted_impurity_temp = p_left * gini_calc(left_labels) + (1-p_left) * gini_calc(right_labels)
                #print('local_imp: %.4f \n' % weighted_impurity_temp)
        
                #set new feature index if impurity improved
                if(weighted_impurity_temp <= weighted_impurity):
                    weighted_impurity = weighted_impurity_temp
                    split_val = split_feature[split]
                    split_index = split
                
        wimps.append(weighted_impurity)
        split_vals.append(split_val) 
        split_indices.append(split_index)
        sort_indices.append(sort_index)
    
        
    minImp = np.argmin(wimps)
    #print(minImp)
    split_val = split_vals[minImp]
    split_index = split_indices[minImp]
    sort_index = sort_indices[minImp]
    
    left_features = (node_features[sort_index])[:split_index+1,:]
    right_features = (node_features[sort_index])[split_index+1:,:]
    left_labels = (node_labels[sort_index])[:split_index+1]
    right_labels = (node_labels[sort_index])[split_index+1:]
        
    
    #print('split_feature', minImp)
    #print('gini', np.min(wimps))
    #print('left_labels', left_labels)
    #print('right_labels', right_labels) 
    
    return split_val, minImp, left_features, left_labels, right_features, right_labels

class Node():
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels 
        self.children = []
        self.split = 0.5
        self.split_index = 20
        
    def isPure(self):
        return np.min(self.labels) == np.max(self.labels)
    
def create_tree(node):
    oneList = []
    if(node.isPure()):
        return        
    else:
        
        split_val, split_feat_index, left_features, left_labels, right_features, right_labels = best_split(node.features, node.labels)
        node.split = split_val
        node.split_index = split_feat_index
        
        print(featNames[node.split_index])
        
        left_child = Node(left_features, left_labels)
        right_child = Node(right_features, right_labels)
        node.children.append(left_child)
        node.children.append(right_child)
        
        create_tree(left_child)
        create_tree(right_child)
```

```python
root = Node(train_features,train_labels)
create_tree(root)
```

```python
def traverse(root,names):
    current_level = [root]
    while current_level:
        temp = ''
        i=0
        for node in current_level:
            i += 1
            lbs, cts = np.unique(node.labels, return_counts=True)
            temp += ' ' + names[node.split_index] + ' ' + str(node.split) + ' ' + str(cts) + str(lbs)
            if(i%2==0):
                temp += '    '
        print(temp)
        next_level = list()
        for n in current_level:
            if not n.isPure():
                next_level.append(n.children[0])
                next_level.append(n.children[1])
                
        current_level = next_level
traverse(root, featNames)
```

```python
def predict(node, inp):
    if(node.isPure()):
        return node.labels[0]
    else:
        if(inp[node.split_index] <= node.split):
            pred = predict(node.children[0], inp)
        else:
            pred = predict(node.children[1], inp)
        return pred
def pred_samples(root, inp):
    res = []
    for i in range(inp.shape[0]):
        res.append(predict(root, inp[i,:]))
    return np.asarray(res)
```

```python
def predictprint(node, inp, names):
    if(node.isPure()):
        return node.labels[0]
    else:
        print(names[node.split_index])
        if(inp[node.split_index] <= node.split):
            pred = predictprint(node.children[0], inp, names)
        else:
            pred = predictprint(node.children[1], inp, names)
        return pred
```

```python
def traintestresults(root, train_features, train_labels, test_features, test_labels):
    #TRAINING ACCURACIES
    train_pred = pred_samples(root, train_features)
    #print(train_pred.shape, train_pred.min(), train_pred.max())
    print('Training Accuracy: %.3f%%' % (np.sum(train_pred==train_labels)
          /train_pred.shape[0]*100))
    numSamples = np.zeros(3)
    truePreds =  np.zeros(3)
    for sample, result in zip(train_labels, train_pred):
        numSamples[int(result)-1] += 1
        if(sample == result):
            truePreds[int(result)-1] += 1
    classAccs = truePreds/numSamples*100
    print('Class-Based Accuracies:')
    for i in range(3):
        print('Accuracy of Class %d: %.3f%%' % (i+1,classAccs[i]))
    cm = confusion_matrix(train_labels, train_pred, labels=[1,2,3])
    cm_display = ConfusionMatrixDisplay(cm, display_labels=[1,2,3]).plot()
    plt.title('Confusion Matrix for Training Data')
    plt.show()
    
    #TEST ACCURACIES
    test_pred = pred_samples(root, test_features)
    #print(test_pred.shape, test_pred.min(), test_pred.max())
    print('Test Accuracy: %.3f%%' % (np.sum(test_pred==test_labels)
          /test_pred.shape[0]*100))
    numSamples = np.zeros(3)
    truePreds =  np.zeros(3)
    for sample, result in zip(test_labels, test_pred):
        numSamples[int(result)-1] += 1
        if(sample == result):
            truePreds[int(result)-1] += 1
    classAccs = truePreds/numSamples*100
    print('Class-Based Accuracies:')
    for i in range(3):
        print('Accuracy of Class %d: %.3f%%' % (i+1,classAccs[i]))
        
    cm = confusion_matrix(test_labels, test_pred, labels=[1,2,3])
    cm_display = ConfusionMatrixDisplay(cm, display_labels=[1,2,3]).plot()
    plt.title('Confusion Matrix for Test Data')
    plt.show()
traintestresults(root, train_features, train_labels, test_features, test_labels)
```
