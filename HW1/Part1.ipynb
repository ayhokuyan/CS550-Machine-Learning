```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn import tree
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
import os
```

```python
#set current working directory
cwd = os.getcwd()
#load train data as csv
train_data = pd.read_csv(os.path.join(cwd, 'train_data.txt'), sep=' ', header=None)
train_data.reset_index(drop=True, inplace=True)
train_data = train_data.to_numpy()[:,:-2]

#load test data as csv
test_data = pd.read_csv(os.path.join(cwd, 'test_data.txt'), sep=' ', header=None)
test_data.reset_index(drop=True, inplace=True)
test_data = test_data.to_numpy()[:,:-2]
```

```python
#load class names 
featNames = pd.read_csv(os.path.join(cwd, 'thy_cost.txt'), sep=':', header=None)
#featNames.reset_index(drop=True, inplace=True)
featNames = featNames.to_numpy()[:,0]
print(featNames, featNames.shape)
featNames = np.append(featNames, 'extracted')
```

```python
print(train_data.shape)
print(test_data.shape)
numFeatures = train_data.shape[1]
train_features = train_data[:,:numFeatures-1]
train_labels = train_data[:,numFeatures-1]
test_features = test_data[:,:numFeatures-1]
test_labels = test_data[:,numFeatures-1]
```

```python
def traintestresults(clf, train_features, train_labels, test_features, test_labels):
    #TRAINING ACCURACIES
    train_pred = clf.predict(train_features)
    #print(train_pred.shape, train_pred.min(), train_pred.max())
    print('Training Accuracy: %.3f%%' % (np.sum(train_pred==train_labels)
          /train_pred.shape[0]*100))
    numSamples = np.zeros(3)
    truePreds =  np.zeros(3)
    for sample, result in zip(train_labels, train_pred):
        numSamples[int(result)-1] += 1
        if(sample == result):
            truePreds[int(result)-1] += 1
    classAccs = truePreds/numSamples*100
    print('Class-Based Accuracies:')
    for i in range(3):
        print('Accuracy of Class %d: %.3f%%' % (i+1,classAccs[i]))
    cm = confusion_matrix(train_labels, train_pred, labels=[1,2,3])
    cm_display = ConfusionMatrixDisplay(cm, display_labels=[1,2,3]).plot()
    plt.title('Confusion Matrix for Training Data')
    plt.show()
    
    #TEST ACCURACIES
    test_pred = clf.predict(test_features)
    #print(test_pred.shape, test_pred.min(), test_pred.max())
    print('Test Accuracy: %.3f%%' % (np.sum(test_pred==test_labels)
          /test_pred.shape[0]*100))
    numSamples = np.zeros(3)
    truePreds =  np.zeros(3)
    for sample, result in zip(test_labels, test_pred):
        numSamples[int(result)-1] += 1
        if(sample == result):
            truePreds[int(result)-1] += 1
    classAccs = truePreds/numSamples*100
    print('Class-Based Accuracies:')
    for i in range(3):
        print('Accuracy of Class %d: %.3f%%' % (i+1,classAccs[i]))
        
    cm = confusion_matrix(test_labels, test_pred, labels=[1,2,3])
    cm_display = ConfusionMatrixDisplay(cm, display_labels=[1,2,3]).plot()
    plt.title('Confusion Matrix for Test Data')
    plt.show()
```

```python
#BASE CASE
clf = tree.DecisionTreeClassifier(random_state=24, criterion='entropy')
clf.fit(train_features, train_labels)

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=300)
tree.plot_tree(clf, feature_names=featNames,filled = True, precision=3);
plt.show()

traintestresults(clf, train_features, train_labels, test_features, test_labels)
```

```python
#PRUNING
clf = tree.DecisionTreeClassifier(random_state=24, criterion='entropy', min_impurity_decrease=0.005)
#clf = tree.DecisionTreeClassifier(random_state=24, criterion='entropy', max_depth=5)
clf.fit(train_features, train_labels)

fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=300)
tree.plot_tree(clf, feature_names=featNames,filled = True, precision=3);
#fig.savefig("decistion_tree.png")
```

```python
traintestresults(clf, train_features, train_labels, test_features, test_labels)
```

```python
means = np.mean(train_features, axis=0)
print(means)
stds = np.std(train_features, axis=0)
print(stds)
train_features_nor = (train_features-means)/stds
test_features_nor = (test_features-means)/stds
```

```python
clf = tree.DecisionTreeClassifier(random_state=24)
clf.fit(train_features_nor, train_labels)


fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=300)
tree.plot_tree(clf, feature_names=featNames,filled = True);
plt.show()

traintestresults(clf, train_features_nor, train_labels, test_features_nor, test_labels)
```

```python
(unique, counts) = np.unique(train_labels, return_counts=True)
frequencies = np.asarray((unique, counts))
numClasses = 3
#threshold the training samples on the lowest class count. 
minThr = np.min(counts) 
train_features_red = []
train_labels_red = []
for i in range(numClasses):
    #for reproducability
    np.random.seed(0)
    samples = train_features[train_labels == i+1]
    samples = samples[np.random.permutation(samples.shape[0])][:minThr,:]
    train_features_red.append(np.asarray(samples))
    train_labels_red.append(minThr * [i+1])
train_features_red = np.vstack(np.asarray(train_features_red))
train_labels_red = np.hstack(np.asarray(train_labels_red))
```

```python
clf = tree.DecisionTreeClassifier(random_state=0)
clf.fit(train_features_red, train_labels_red)


fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (10,10), dpi=300)
tree.plot_tree(clf, feature_names=featNames,filled = True);
plt.show()

traintestresults(clf, train_features_red, train_labels_red, test_features, test_labels)
```
